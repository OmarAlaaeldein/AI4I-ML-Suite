```python
import pandas as pd

# Read the JSON file
df = pd.read_json('results.json')

# Drop duplicates based on the score column
df_unique = df.drop_duplicates(subset='score', keep='first')

df_unique[["f1","hamming_loss","auc","precision","accuracy","recall","score"]]=df_unique[["f1","hamming_loss","auc","precision","accuracy","recall","score"]]*100

# round the values to 2 decimal places
df_unique[["f1","hamming_loss","auc","precision","accuracy","recall","score"]] = df_unique[["f1","hamming_loss","auc","precision","accuracy","recall","score"]].round(2)

# Get the top 10 unique results based on score
top_10_results = df_unique.nlargest(10, 'score')

# get the top score for each classifier
top_classifiers = df_unique.loc[df_unique.groupby('classifier')['score'].idxmax()]
# remove hyperparameters column
#top_classifiers = top_classifiers.drop(columns=['hyperparameters'])
# sort the classifiers based on score
top_classifiers = top_classifiers.sort_values(by='score', ascending=False)

# add "%" to results
top_10_results[["f1","hamming_loss","auc","precision","accuracy","recall","score"]] = top_10_results[["f1","hamming_loss","auc","precision","accuracy","recall","score"]].astype(str) + '\%'

top_classifiers[["f1","hamming_loss","auc","precision","accuracy","recall","score"]] = top_classifiers[["f1","hamming_loss","auc","precision","accuracy","recall","score"]].astype(str) + '\%'
```
