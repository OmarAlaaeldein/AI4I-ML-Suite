```python
import pandas as pd

# Read the JSON file
df = pd.read_json('results.json')

# Drop duplicates based on the score column
df_unique = df.drop_duplicates(subset='score', keep='first')

# Get the top 10 unique results based on score
top_10_results = df_unique.nlargest(10, 'score')

# get the top score for each classifier
top_classifiers = df_unique.loc[df_unique.groupby('classifier')['score'].idxmax()]
```
